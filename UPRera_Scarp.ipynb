```python
#Creating the Empty List of the All Attributes..
Project_Name = []
Reg_no = []
Distict =[]
Start_Date =[]
End_Date = []
```

```python
#Project_names
for i in range(2,1338):
    x_path = '//*[@id="grdPojDetail_ctl{}_lblProjectName"]'
    current_path = x_path.format(str(i).zfill(2))
    x_path_data = driver.find_element_by_xpath(current_path)
    Project_Name.append(x_path_data.text)

    x_path = '//*[@id="grdPojDetail_ctl{}_lblRegistrationNo"]'
    current_path = x_path.format(str(i).zfill(2))
    x_path_data = driver.find_element_by_xpath(current_path)
    Reg_no.append(x_path_data.text)

    x_path = '//*[@id="grdPojDetail_ctl{}_lblDistrict"]'
    current_path = x_path.format(str(i).zfill(2))
    x_path_data = driver.find_element_by_xpath(current_path)
    Distict.append(x_path_data.text)

    x_path = '//*[@id="grdPojDetail_ctl{}_lblStartDate"]'
    current_path = x_path.format(str(i).zfill(2))
    x_path_data = driver.find_element_by_xpath(current_path)
    Start_Date.append(x_path_data.text)

    x_path = '//*[@id="grdPojDetail_ctl{}_lblEndDate"]'
    current_path = x_path.format(str(i).zfill(2))
    x_path_data = driver.find_element_by_xpath(current_path)
    End_Date.append(x_path_data.text)

    

    
```

```python
Number_of_projects = []
complaints_against_the_promoter = []
complaints_to_this_Project = []
```

```python
title_window = driver.current_window_handle
```

```python
for i in range(1,1338):
    c = f'''/html/body/form/div[4]/section/div/div[3]/div/div/div[3]/div/div/table/tbody/tr[{i}]/td[9]/a/i'''
    click = driver.find_element_by_xpath(c)
    click.click()
    WebDriverWait(driver, 1000).until(EC.number_of_windows_to_be(2))
    print("NEW WINDOW")
    all_windows = driver.window_handles  #  window_handle means the new tab open .
    driver.switch_to.window(all_windows[1]) 
    # WebDriverWait(driver, 1000).until(EC.number_of_windows_to_be(2))
    try:
        WebDriverWait(driver, 200).until(EC.presence_of_element_located((By.XPATH, '/html/body/form/div[3]/section[2]/div/div/div/div/div[2]/div[5]/div/div/div[2]/div/div/div/div[1]/div[2]')))
        x_path = '/html/body/form/div[3]/section[2]/div/div/div/div/div[2]/div[5]/div/div/div[2]/div/div/div/div[1]/div[2]'
        x_path_data = driver.find_element_by_xpath(x_path)
        Number_of_projects.append(x_path_data.text)
    
        x_path = '/html/body/form/div[3]/section[2]/div/div/div/div/div[2]/div[5]/div/div/div[2]/div/div/div/div[2]/div[2]'
        x_path_data = driver.find_element_by_xpath(x_path)
        complaints_against_the_promoter.append(x_path_data.text)
    
        x_path = '/html/body/form/div[3]/section[2]/div/div/div/div/div[2]/div[5]/div/div/div[2]/div/div/div/div[3]/div[2]'
        x_path_data = driver.find_element_by_xpath(x_path)
        complaints_to_this_Project.append(x_path_data.text)
    except Exception as e:
        print("ERROR FOUND AT" , i , "=====>" , e)
        Number_of_projects.append(None)
        complaints_against_the_promoter.append(None)
        complaints_to_this_Project.append(None)

        
    ##click the view button
    ##it will go to 2nd page 
 
     # switch to tab
    ##do stuff on 2nd page
    
    

    print(i , "page done")
    
    
    ##close 2nd page
    driver.close()  # close to tab 
    driver.switch_to.window(title_window) #return to main page
```
